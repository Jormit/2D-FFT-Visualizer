```python
import numpy as np
import matplotlib.pyplot as plt

freq_x = 4
freq_y = 1
a = np.zeros((100, 100))
freq = 10
for x_i in range(100):
    for y_i in range(100):
        a[y_i, x_i] = np.cos(2*np.pi*(x_i*freq_x + y_i*freq_y)/100)
```

```python
plt.figure(dpi=100)
plt.imshow(a)
plt.title("Spatial Domain")
plt.xlabel("x")
plt.ylabel("y")
plt.figure(plt.figure(dpi=100))
plt.imshow(np.abs(np.fft.fftshift(np.fft.fft2(a))))
plt.title("Frequency Domain");
```
